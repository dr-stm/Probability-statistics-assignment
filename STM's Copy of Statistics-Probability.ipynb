5. A gourmet cafe serves exotic salmon at lunch and dinner with the number of customers ordering them as follows:

Lunch:          0   1   2

Probability     0.3 0.5 0.2


Dinner:          0   1   2

Probability     0.2 0.4 0.4

What is the expected total demand for the fish in a day?

```python
# Given the above distribution, the joint distribution will be as follows:

joint_dist= [[0.06, 0.10, 0.04, 0.2], 
               [0.12, 0.20, 0.08, 0.4], \
               [0.12, 0.20, 0.08, 0.4], \
               [0.3, 0.5, 0.2]]

joint_distDf = pd.DataFrame(joint_dist, 
                                  columns=["Lunch = 0", "Lunch = 1", \
                                           "Lunch = 2", "Total"])

joint_distDf.index = ["Dinner = 0", "Dinner = 1", "Dinner = 2", "Total"]

joint_distDf
```

```python
# Sequel to the joint distribution table above, the joint demand and the probability will be as follows

joint_demand = {"Joint demand":["0", "1", "2", "3", "4"], \
                "Probability fraction":[0.06, 0.22, 0.36, 0.28, 0.08], \
                "Probability percent": [6, 22, 36, 28, 8]};

joint_demandDf = pd.DataFrame(data = joint_demand)

joint_demandDf
```

```python
#The graphical representation will be as follows:

joint_demandDf.plot.bar(x="Joint demand", y="Probability percent", rot = 0, fontsize = 12, \
                        title="Bar chart of the joint demand probability distribution", \
                        color = "#e9868b", ylabel = "Probability in percent")
```
